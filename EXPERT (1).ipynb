```python
# Knowledge base mapping symptoms to potential medical conditions
knowledge_base = {
    "fever": ["Flu", "COVID-19", "Common Cold"],
    "cough": ["Flu", "COVID-19", "Bronchitis"],
    "headache": ["Migraine", "Tension Headache", "Sinusitis"],
    "sore throat": ["Strep Throat", "Tonsillitis", "Common Cold"],
    "shortness of breath": ["Asthma", "COVID-19", "Pneumonia"],
    "nausea": ["Gastroenteritis", "Food Poisoning", "Pregnancy"],
}

class MedicalExpertSystem:
    def __init__(self):
        self.knowledge_base = knowledge_base

    def check_symptoms(self, symptoms):
        condition_frequency = {}

        for symptom in symptoms:
            if symptom in self.knowledge_base:
                for condition in self.knowledge_base[symptom]:
                    if condition in condition_frequency:
                        condition_frequency[condition] += 1
                    else:
                        condition_frequency[condition] = 1
            else:
                print(f"‚ö†Ô∏è  Warning: '{symptom}' is not in the knowledge base.")

        if not condition_frequency:
            return "No known medical conditions found for the provided symptoms."

        # Sort conditions by how frequently they matched
        sorted_conditions = sorted(condition_frequency.items(), key=lambda x: x[1], reverse=True)
        return sorted_conditions

    def display_results(self, sorted_conditions):
        print("\nüìã Possible medical conditions based on your symptoms:\n")
        for condition, frequency in sorted_conditions:
            print(f"- {condition} (matched {frequency} symptom{'s' if frequency > 1 else ''})")

        print("\nüîç Note: This system provides suggestions only based on known patterns.")
        print("‚öïÔ∏è Please consult a licensed medical professional for an accurate diagnosis.")

if __name__ == "__main__":
    expert_system = MedicalExpertSystem()

    print("ü©∫ Welcome to the Medical Expert System!")
    print("Please enter your symptoms separated by commas (e.g., fever, cough):")
    
    user_input = input("Symptoms: ")
    symptoms = [symptom.strip().lower() for symptom in user_input.split(",") if symptom.strip()]

    if not symptoms:
        print("‚ùå Error: No symptoms entered.")
    else:
        result = expert_system.check_symptoms(symptoms)
        if isinstance(result, str):
            print(result)
        else:
            expert_system.display_results(result)
```
