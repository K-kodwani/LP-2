```python
g = 0  # Global variable to track the number of moves

def print_board(elements):
    
    for i in range(9):
        if i % 3 == 0:
            print()
        if elements[i] == -1:
            print("_", end=" ")
        else:
            print(elements[i], end=" ")
    print()

def solvable(start):

    inv = 0
    for i in range(9):
        if start[i] <= 1:
            continue
        for j in range(i + 1, 9):
            if start[j] == -1:
                continue
            if start[i] > start[j]:
                inv += 1
    return inv % 2 == 0  # Solvable if inversion count is even

def heuristic(start, goal):

    global g
    h = 0
    for i in range(9):
        for j in range(9):
            if start[i] == goal[j] and start[i] != -1:
                h += (abs(j - i)) // 3 + (abs(j - i)) % 3  # Manhattan distance
    return h + g

def moveleft(start, position):
   
    start[position], start[position - 1] = start[position - 1], start[position]

def moveright(start, position):
 
    start[position], start[position + 1] = start[position + 1], start[position]

def moveup(start, position):
   
    start[position], start[position - 3] = start[position - 3], start[position]

def movedown(start, position):
    
    start[position], start[position + 3] = start[position + 3], start[position]

def movetile(start, goal):
    
    emptyat = start.index(-1)  # Find empty tile position
    row, col = emptyat // 3, emptyat % 3  # Find row and column

    # Copies of start state for different move directions
    t1, t2, t3, t4 = start[:], start[:], start[:], start[:]
    f1, f2, f3, f4 = 100, 100, 100, 100  # Initialize heuristic values

    if col - 1 >= 0:  # Left move possible
        moveleft(t1, emptyat)
        f1 = heuristic(t1, goal)

    if col + 1 < 3:  # Right move possible
        moveright(t2, emptyat)
        f2 = heuristic(t2, goal)

    if row + 1 < 3:  # Down move possible
        movedown(t3, emptyat)
        f3 = heuristic(t3, goal)

    if row - 1 >= 0:  # Up move possible
        moveup(t4, emptyat)
        f4 = heuristic(t4, goal)

    # Choose the move with the lowest heuristic value
    min_heuristic = min(f1, f2, f3, f4)

    if f1 == min_heuristic:
        moveleft(start, emptyat)
    elif f2 == min_heuristic:
        moveright(start, emptyat)
    elif f3 == min_heuristic:
        movedown(start, emptyat)
    elif f4 == min_heuristic:
        moveup(start, emptyat)

def solveEight(start, goal):

    global g
    g += 1  # Increment move count
    movetile(start, goal)
    print_board(start)
    
    f = heuristic(start, goal)
    
    if f == g:  # If heuristic value matches move count, it's solved
        print("Solved in {} moves".format(f))
        return
    solveEight(start, goal)  # Recursive call

def main():
  
    global g
    start = []
    goal = []
    
    print("Enter the start state (Use -1 for empty tile):")
    for i in range(9):
        start.append(int(input()))
    
    print("Enter the goal state (Use -1 for empty tile):")
    for i in range(9):
        goal.append(int(input()))

    print("\nInitial State:")
    print_board(start)

    if solvable(start):
        solveEight(start, goal)
        print("Solved in {} moves".format(g))
    else:
        print("Not possible to solve")

if __name__ == '__main__':
    main()

print('Name:Aditya Chougale\nRoll no: 22127')
```
