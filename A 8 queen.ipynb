```python
#!/usr/bin/env python
# coding: utf-8

# In[2]:


N = 8
g = 0  # Move counter

def print_board(board):
    for row in board:
        print(" ".join("Q" if col else "_" for col in row))
    print()

def is_safe(board, row, col, left_row, upper_diag, lower_diag):
    if left_row[row] or upper_diag[row + col] or lower_diag[row - col + N - 1]:
        return False
    return True

def solve(board, col, left_row, upper_diag, lower_diag):
    global g
    if col >= N:
        return True  
    
    # Try placing the queen in a more optimal order (middle rows first)
    order = [3, 4, 2, 5, 1, 6, 0, 7]
    
    for i in order:
        if is_safe(board, i, col, left_row, upper_diag, lower_diag):
            board[i][col] = 1
            left_row[i] = upper_diag[i + col] = lower_diag[i - col + N - 1] = True
            g += 1  # Increment move count
            print(f"\nState after placing Queen at ({i}, {col}):")
            print_board(board)

            if solve(board, col + 1, left_row, upper_diag, lower_diag):  
                return True  # Stop once a solution is found
            
            # Backtrack
            board[i][col] = 0
            left_row[i] = upper_diag[i + col] = lower_diag[i - col + N - 1] = False
    
    return False

def solve_n_queens():
    global g
    board = [[0] * N for _ in range(N)]  
    left_row = [False] * N
    upper_diag = [False] * (2 * N - 1)
    lower_diag = [False] * (2 * N - 1)

    print("Initial Empty Board:")
    print_board(board)

    if not solve(board, 0, left_row, upper_diag, lower_diag):
        print("No solution exists")
    else:
        print("\nFinal Solution:")
        print_board(board)  

    print(f"Total Moves: {g}")  # Moves at the end

if __name__ == "__main__":
    solve_n_queens()

print('Name: Aditya Chougale\nRoll no: 22127')


# In[ ]:




```
