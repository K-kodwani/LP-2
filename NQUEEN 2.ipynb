```python
print("Name: Hitesh Pal\n", "Roll No. TACO22132")

class NQueens:
    def __init__(self, n):
        self.n = n
        self.board = [['.' for _ in range(n)] for _ in range(n)]
        self.solutions = []
        
        # Branch and Bound optimization
        self.columns = [False] * n
        self.diag1 = [False] * (2 * n - 1)  # row - col + n - 1
        self.diag2 = [False] * (2 * n - 1)  # row + col

    def solve(self, row=0):
        if row == self.n:
            self.solutions.append([''.join(r) for r in self.board])
            return

        for col in range(self.n):
            d1 = row - col + self.n - 1
            d2 = row + col
            if not self.columns[col] and not self.diag1[d1] and not self.diag2[d2]:
                # Place queen
                self.board[row][col] = 'Q'
                self.columns[col] = self.diag1[d1] = self.diag2[d2] = True

                self.solve(row + 1)

                # Backtrack
                self.board[row][col] = '.'
                self.columns[col] = self.diag1[d1] = self.diag2[d2] = False

    def print_solutions(self):
        if not self.solutions:
            print("No solutions found.")
        else:
            print(f"\nTotal solutions: {len(self.solutions)}\n")
            for solution in self.solutions:
                for row in solution:
                    print(row)
                print()

# Main Execution
if __name__ == '__main__':
    print("\nName: Hitesh Pal\nRoll no: TACO22132")
    
    input_n = int(input("Enter a number: "))  # User enters a number
    n = 2 ** input_n                         # n becomes 2^input_n
    print(f"\nChessboard size will be {n} x {n}\n")

    queens = NQueens(n)
    queens.solve()
    queens.print_solutions()
```
